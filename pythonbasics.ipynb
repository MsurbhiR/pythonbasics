Que 8 :  Write a program to demonstrate the use of logical operators.


```python
x = 10
y = 5
z = 20

print("Logical AND:")
print(f"x = {x}, y = {y}, z = {z}")
print(f"x > y and z > y: {x > y and z > y}") # Both conditions must be True
print(f"x < y and y < z: {x < y and y < z}") # One condition is False

print("\nLogical OR:")
print(f"x = {x}, y = {y}, z = {z}")
print(f"x > y or z > y: {x > y or z > y}")  # At least one condition must be True
print(f"x < y or y > z: {x < y or y > z}")  # One condition is False, but other is True

print("\nLogical NOT:")
print(f"x = {x}, y = {y}")
print(f"not (x > y): {not (x > y)}")  # Reverses the result of the condition
print(f"not (y > x): {not (y > x)}")  # Reverses the result of the condition
```

Que 13 :  Write a Python program to find the sum of all even numbers between 1 and 50.

```python
n = 50
sum = 0
for i in range(1,n+1):
  if i % 2 == 0:
    sum = sum + i
print(sum)
```

Que 14 :  Write a program to reverse a string using a while loop.

```python
str = input("enter a string : ")
i = len(str) - 1
while i >= 0:
  print(str[i],end="")
  i = i - 1
```

Que 15 :  Write a Python program to calculate the factorial of a number provided by the user using a while loop.


```python
n = int(input("enter a number : "))
fact = 1
while n > 0:
  fact = fact * n
  n = n - 1
print(fact)
```
