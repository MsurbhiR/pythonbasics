Que 2 :  Explain the difference between mutable and immutable data types with examples.
  - Mutable data types are those that allow modification of their content without changing their identity.
   - list, dictionary & set is mutable data types.



```python
#example of mutable data types
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)
```

-  immutable data types not allow any modification once after created.
- tuples , strings & number are immutable data types.

```python
#example of immutable data types
my_tuple = (1, 2, 3)
my_tuple[0] = 10
print(my_tuple)
```
